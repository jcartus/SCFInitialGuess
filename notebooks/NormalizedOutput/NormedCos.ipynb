```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow import keras
import tensorflow as tf

plt.style.use("seaborn")
```

```python
n = 1000
x = np.random.randn(n) * 2 * np.pi + np.pi
y = np.cos(x)
```

# Not normed 

```python
ind = int(0.8 * n)
def split(t):
    return t[:ind], t[ind:]

def norm(t):
    mu = np.mean(t)
    std = np.std(t)
    return (t - mu) / std, mu, std

x_norm, x_mu, x_std = norm(x)
    
x_train, x_test = split(x_norm)
y_train, y_test = split(y)
```

```python
plt.scatter(x_train, y_train, label="train")
plt.scatter(x_test, y_test, label="test")

plt.legend()
```

```python
def make_model(activation="elu"):
    model = keras.Sequential()
    model.add(keras.layers.Dense(5, activation=activation, input_dim=1))
    model.add(keras.layers.Dense(5, activation=activation))
    model.add(keras.layers.Dense(5, activation=activation))
    model.add(keras.layers.Dense(1))
    
    model.compile(
        optimizer=tf.train.AdamOptimizer(1e-2),
        loss='mse'
    )
    
    return model
```

```python
keras.backend.clear_session()

model = make_model()

history = model.fit(
    x_train,
    y_train,
    validation_data=(x_test,y_test),
    epochs=500
)
```

```python
plt.semilogy(history.history["loss"])
plt.semilogy(history.history["val_loss"])
```

```python
plt.scatter(x_train, y_train)
plt.plot(np.sort(x_test), model.predict(np.sort(x_test)))
```
