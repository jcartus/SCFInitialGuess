```python
from functools import reduce
keys = "a,b,c,d".split(",")
data = {}
for k in keys:
    data[k] = np.random.randint(low=6, high=16, size=(20))

%matplotlib inline

def iterations_histogram(data, bin_boundaries=None):
    """
    data ... dict label: data
    bin_boundaries ... tupel lower (inclusive), upper(noninclusive)
    """
    
    if bin_boundaries is None:
        values = np.array(list(data.values()))
        bin_boundaries = (np.min(values), np.max(values) + 2 ) 
    
    bins = np.arange(*bin_boundaries) - 0.5
    default_width=0.8
    width = default_width / len(data)
    
    for i, (label, values) in enumerate(data.items()):
        hists, edges = np.histogram(values, bins=bins)
        
        center = (edges[:-1] + edges[1:]) / 2 + i * width
        position = edges
        plt.bar(center, hists, label=label, width=width)
    
    plt.legend()

iterations_histogram(data)    
plt.show()
```
